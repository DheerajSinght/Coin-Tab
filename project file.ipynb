# finding total weight as per company x in kg

# Generation of Weight Slab

```python
def weight_slab(weight):
    if weight  == 0:
        return weight
    elif weight  > 0.5:
        return int(weight) + 1
    else:
        return int(weight) + 0.5
```

```python
df2['Weight_Slab_As_Per_x_KG']=df2['Total_Weight_Per_x_kg'].apply(weight_slab)
df_cc['Weight_Slab_Charged_By_Courier_Company_kg']=(df_cc['Total_Weight_As_Per_Courier_Company_(KG)']).apply(weight_slab)
```

# total expected charge

```python
total_expected_charge = []

for i in range(len(df2)):
    delivery_zone = df2['Delivery_zone_as_per_x'][i]
    fwd_category = 'fwd_' + delivery_zone
    fwd_fixed = df_r[fwd_category + '_fixed'][0]
    fwd_additional = df_r[fwd_category + '_additional'][0]
    rto_category = 'rto_' + delivery_zone
    rto_fixed = df_r[rto_category + '_fixed'][0]
    rto_additional = df_r[rto_category + '_additional'][0]
    weight_slab = df2['Weight_Slab_As_Per_x_KG'][i]
    shipment_type = df2['Type_of_Shipment'][i]

    if shipment_type == 'Forward charges':
        if weight_slab <= 0.5:
            total_expected_charge.append(fwd_fixed)
        else:
            additional_weight = (weight_slab - 0.5) / 0.5
            total_expected_charge.append(fwd_fixed + additional_weight * fwd_additional)
    elif shipment_type == 'Forward and RTO charges':
        if weight_slab <= 0.5:
            total_expected_charge.append(fwd_fixed + rto_fixed)
        else:
            additional_weight = (weight_slab - 0.5) / 0.5
            total_expected_charge.append(fwd_fixed + additional_weight * (fwd_additional + rto_additional))

df2['Expected_Charge_As_Per_x_Rs'] = total_expected_charge
```

# final result 2 (summary)

```python
# Calculate correctly charged orders
correctly_charged = output2[output2['Difference Between Expected Charges and Billed Charges (Rs)'] == 0.0]
total_correctly_charged = len(correctly_charged)
total_correct_charges = sum(correctly_charged['Charges Billed By Courier Company (Rs)'])

# Calculate over charged orders
over_charged = output2[output2['Difference Between Expected Charges and Billed Charges (Rs)'] < 0.0]
total_over_charged = len(over_charged)
total_over_charges = np.absolute(sum(over_charged['Difference Between Expected Charges and Billed Charges (Rs)']))

# Calculate under charged orders
under_charged = output2[output2['Difference Between Expected Charges and Billed Charges (Rs)'] > 0.0]
total_under_charged = len(under_charged)
total_under_charges = np.absolute(sum(under_charged['Difference Between Expected Charges and Billed Charges (Rs)']))

# Create summary data
correctly = ["Total Orders where x has been correctly charged", total_correctly_charged, total_correct_charges]
over = ["Total Orders where x has been over charged", total_over_charged, total_over_charges]
under = ["Total Orders where x has been under charged", total_under_charged, total_under_charges]
```

```python
result2=pd.DataFrame([correctly,over,under],columns=['Description','Count','Amount (Rs.)'])
```
